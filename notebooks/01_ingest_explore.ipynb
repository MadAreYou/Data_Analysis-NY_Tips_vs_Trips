## TLC 2024 data collection

This step downloads TLC 2024 monthly parquet files to X:, (X = to remote S3 type  of cloud storage - (iDrive E2)) skips existing files, and validates a sample column.

```python
import os, hashlib, requests
import pandas as pd
from tqdm import tqdm

# TLC Trip Data Downloader for 2024

BASE_URL = "https://d37ci6vzurychx.cloudfront.net/trip-data"
DEST_DIR = r"X:\data\raw\tlc\2024"
YEAR = "2024"

os.makedirs(DEST_DIR, exist_ok=True)

def sha256_checksum(file_path, chunk_size=8192):
    """Compute SHA256 checksum of a file."""
    sha = hashlib.sha256()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(chunk_size), b""):
            sha.update(chunk)
    return sha.hexdigest()

# --- DOWNLOAD LOOP ---
for m in range(1, 13):
    month = f"{m:02d}"
    file_name = f"yellow_tripdata_{YEAR}-{month}.parquet"
    url = f"{BASE_URL}/{file_name}"
    dest_path = os.path.join(DEST_DIR, file_name)

    if os.path.exists(dest_path):
        print(f"✔ File already exists: {file_name}")
        continue

    print(f"⬇ Downloading {file_name} ...")
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(dest_path, "wb") as f:
            for chunk in tqdm(r.iter_content(chunk_size=8192), desc=file_name):
                if chunk:
                    f.write(chunk)

    checksum = sha256_checksum(dest_path)
    print(f"✔ Completed {file_name}, SHA256: {checksum}")

    # --- quick validation read ---
    try:
        df_sample = pd.read_parquet(dest_path, engine="pyarrow", columns=["tpep_pickup_datetime"])
        print(f"Sample rows: {len(df_sample)}\n")
    except Exception as e:
        print(f"⚠ Could not read sample from {file_name}: {e}")
```

## NOAA 2024 Weather Data Ingestion & City-wide Hourly Aggregation


This step downloads NOAA station CSVs for 2024:, computes checksums, and handles errors. Cleans station CSVs to NYC local hour, aggregates to citywide hourly metrics, writes locally, then copies to X:.

```python
# --- NOAA WEATHER DATA INGESTION PLAN ---

import os, hashlib, requests
from tqdm import tqdm

BASE_INDEX = "https://www.ncei.noaa.gov/data/local-climatological-data/access/2024"
DEST_DIR = r"X:\data\raw\noaa\2024"
os.makedirs(DEST_DIR, exist_ok=True)

# Map human-friendly station names to the correct USAF-WBAN filenames
STATIONS = {
    "JFK":        "72501504789.csv",
    "LaGuardia":  "72503014732.csv",
    "CentralPark":"72505394728.csv",
    "Newark":     "72502014734.csv",
    "Teterboro":  "72502594741.csv",
}

def sha256_checksum(file_path, chunk_size=1024 * 1024):
    sha = hashlib.sha256()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(chunk_size), b""):
            sha.update(chunk)
    return sha.hexdigest()

for name, fname in STATIONS.items():
    url = f"{BASE_INDEX}/{fname}"
    dest_path = os.path.join(DEST_DIR, f"station_{name}_2024.csv")

    if os.path.exists(dest_path) and os.path.getsize(dest_path) > 0:
        print(f"✔ {name}: already exists ({dest_path})")
        continue

    print(f"⬇ Downloading {name} → {fname}")
    try:
        with requests.get(url, stream=True, timeout=60) as r:
            r.raise_for_status()
            with open(dest_path, "wb") as f:
                for chunk in tqdm(r.iter_content(chunk_size=1024 * 1024), desc=name):
                    if chunk:
                        f.write(chunk)
    except requests.HTTPError as e:
        print(f"❌ HTTP error for {name} ({url}): {e}")
        if os.path.exists(dest_path):
            os.remove(dest_path)
        continue
    except Exception as e:
        print(f"❌ Error for {name}: {e}")
        if os.path.exists(dest_path):
            os.remove(dest_path)
        continue

    # checksum
    checksum = sha256_checksum(dest_path)
    size_mb = round(os.path.getsize(dest_path) / (1024 * 1024), 2)
    print(f"✔ {name}: downloaded {size_mb} MB, SHA256={checksum}")
```

This step probes X: drive write capability and disk info to diagnose network I/O issues.

```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
from statistics import mode
from tqdm import tqdm
import pyarrow
from pathlib import Path
import shutil, time

RAW_PATH = r"X:\data\raw\noaa\2024"
# Primary (X:) plus local canonical output in repo
OUT_PRIMARY = r"X:\data\interim\noaa_hourly_citywide_2024.parquet"
OUT_LOCAL = os.path.abspath(os.path.join("..", "data", "interim", "noaa_hourly_citywide_2024.parquet"))

# Ensure local directory exists
os.makedirs(os.path.dirname(OUT_LOCAL), exist_ok=True)

STATIONS = {
    "JFK":        "station_JFK_2024.csv",
    "LaGuardia":  "station_LaGuardia_2024.csv",
    "CentralPark":"station_CentralPark_2024.csv",
    "Newark":     "station_Newark_2024.csv",
    "Teterboro":  "station_Teterboro_2024.csv",
}

COLS = [
    "DATE",
    "HourlyDryBulbTemperature",
    "HourlyRelativeHumidity",
    "HourlyWindSpeed",
    "HourlyPrecipitation",
    "HourlyVisibility",
    "HourlyPresentWeatherType"
]

def safe_mode(series):
    try:
        vals = series.dropna()
        return mode(vals) if len(vals) else np.nan
    except Exception:
        return np.nan

def clean_noaa_csv(path):
    df = pd.read_csv(path, usecols=lambda c: c in COLS, low_memory=False)
    # --- safer datetime handling around DST ---
    df["DATE"] = pd.to_datetime(df["DATE"], errors="coerce", utc=True)
    # Convert to local tz; floor to hour with DST rules, then drop tz
    df["DATE"] = (
        df["DATE"]
        .dt.tz_convert("America/New_York")
        .dt.floor("h", ambiguous="NaT", nonexistent="shift_forward")
        .dt.tz_localize(None)
    )
    # Numeric conversions
    num_cols = [
        "HourlyDryBulbTemperature",
        "HourlyRelativeHumidity",
        "HourlyWindSpeed",
        "HourlyPrecipitation",
        "HourlyVisibility"
    ]
    for c in num_cols:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    df["Station"] = os.path.basename(path).split("_")[1]

    agg = (
        df
        .groupby(["Station", "DATE"], dropna=True)
        .agg({
            "HourlyDryBulbTemperature": "mean",
            "HourlyRelativeHumidity": "mean",
            "HourlyWindSpeed": "mean",
            "HourlyPrecipitation": "sum",
            "HourlyVisibility": "mean",
            "HourlyPresentWeatherType": safe_mode
        })
        .reset_index()
        .rename(columns={
            "DATE": "observation_hour_local",
            "HourlyDryBulbTemperature": "temp_c",
            "HourlyRelativeHumidity": "rh_percent",
            "HourlyWindSpeed": "wind_ms",
            "HourlyPrecipitation": "precip_mm",
            "HourlyVisibility": "visibility_km",
            "HourlyPresentWeatherType": "weather_code"
        })
    )
    return agg

# === PROCESS ALL STATIONS ===
frames = []
for name, fname in tqdm(STATIONS.items(), desc="Processing stations"):
    fpath = os.path.join(RAW_PATH, fname)
    if not os.path.exists(fpath):
        print(f"⚠ Missing file: {fpath}")
        continue
    frames.append(clean_noaa_csv(fpath))

if not frames:
    raise RuntimeError("No station files were found. Check RAW_PATH and filenames.")

df_all = pd.concat(frames, ignore_index=True)
print(f"Combined hourly rows (all stations): {len(df_all):,}")

# === CITYWIDE AGGREGATION ===
agg_city = (
    df_all
    .groupby("observation_hour_local", dropna=True)
    .agg({
        "temp_c": "mean",
        "rh_percent": "mean",
        "wind_ms": "mean",
        "precip_mm": "sum",
        "visibility_km": "mean",
        "weather_code": safe_mode
    })
    .reset_index()
)

agg_city = agg_city.sort_values("observation_hour_local").reset_index(drop=True)
print(f"Citywide hourly records: {len(agg_city):,}")

# --- canonical write locally first ---
agg_city.to_parquet(OUT_LOCAL, index=False)
print(f"✅ Saved local canonical file to: {OUT_LOCAL}")

# --- then attempt copy to X: with retries (optional) ---
def copy_with_retries(src, dst, retries=3, delay=2.0):
    for i in range(1, retries+1):
        try:
            os.makedirs(os.path.dirname(dst), exist_ok=True)
            shutil.copy2(src, dst)
            print(f"📤 Copied to X: {dst}")
            return True
        except PermissionError as e:
            print(f"Attempt {i}/{retries}: PermissionError copying to {dst}: {e}")
        except Exception as e:
            print(f"Attempt {i}/{retries}: Error copying to {dst}: {e}")
        time.sleep(delay)
    return False

_ = copy_with_retries(OUT_LOCAL, OUT_PRIMARY)
```

### Merge & Export NOAA + TLC datasets

This step reads monthly TLC parquet files, derives NYC local pickup hour, left-joins hourly NOAA weather, writes outputs locally, and then copies to X:.

```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm

# === CONFIG ===
TLC_DIR = r"X:\data\raw\tlc\2024"
NOAA_PATH = r"X:\data\interim\noaa_hourly_citywide_2024.parquet"
# primary outputs on X: (copied after local write)
OUT_PARQUET = r"X:\data\processed\nyc_2024_trips_weather.parquet"
OUT_SAMPLE = r"X:\data\processed\nyc_2024_trips_weather_sample.csv"
# local canonical outputs in repo (write here first)
LOCAL_PARQUET = os.path.abspath(os.path.join("..","data","processed","nyc_2024_trips_weather.parquet"))
LOCAL_SAMPLE  = os.path.abspath(os.path.join("..","data","processed","nyc_2024_trips_weather_sample.csv"))

# Load NOAA hourly dataset
weather = pd.read_parquet(NOAA_PATH)
weather.rename(columns={"observation_hour_local": "pickup_hour_local"}, inplace=True)
print(f"Loaded weather rows: {len(weather):,}")

# Helper: robustly derive NYC local hour from a datetime-like series
def to_local_hour(series, tz="America/New_York"):
    s = pd.to_datetime(series, errors="coerce")
    # If tz-aware, convert to NY and drop tz; if tz-naive, assume local NY already
    if getattr(s.dt, "tz", None) is not None:
        s = s.dt.tz_convert(tz).dt.tz_localize(None)
    return s.dt.floor("h")

# Prepare list of monthly files
files = sorted([f for f in os.listdir(TLC_DIR) if f.endswith(".parquet")])

merged_parts = []
for file in tqdm(files, desc="Merging monthly TLC files"):
    path = os.path.join(TLC_DIR, file)
    print(f"Processing {file} ...")
    df = pd.read_parquet(
        path,
        columns=[
            "tpep_pickup_datetime",
            "PULocationID",
            "trip_distance",
            "fare_amount",
            "tip_amount",
            "total_amount",
            "payment_type",
        ],
    )
    # Convert pickup to NYC local hour (handles tz-aware/naive safely)
    df["pickup_hour_local"] = to_local_hour(df["tpep_pickup_datetime"])
    df.drop(columns=["tpep_pickup_datetime"], inplace=True)

    # Merge with weather (left join)
    merged = df.merge(weather, on="pickup_hour_local", how="left", indicator=True)
    merged["merge_status"] = np.where(merged["_merge"] == "both", "matched", "no_weather_for_hour")
    merged.drop(columns=["_merge"], inplace=True)

    merged_parts.append(merged)

# Concatenate all months (optional; can be huge)
df_final = pd.concat(merged_parts, ignore_index=True)
print(f"Final merged shape: {df_final.shape}")

# === Save outputs: local first, then copy to X: ===
# Ensure local directory exists
os.makedirs(os.path.dirname(LOCAL_PARQUET), exist_ok=True)
df_final.to_parquet(LOCAL_PARQUET, index=False)
print(f"✅ Saved local merged dataset → {LOCAL_PARQUET}")

# Save 50 k-row CSV sample locally
sample = df_final.sample(n=min(50_000, len(df_final)), random_state=42)
sample.to_csv(LOCAL_SAMPLE, index=False)
print(f"✅ Saved local sample CSV → {LOCAL_SAMPLE}")

# Copy to X: with small retry helper (self-contained)
import shutil, time
def copy_with_retries(src, dst, retries=5, delay=2.0):
    for i in range(1, retries+1):
        try:
            os.makedirs(os.path.dirname(dst), exist_ok=True)
            # Remove existing to avoid read-only/lock rename issues
            if os.path.exists(dst):
                try:
                    os.remove(dst)
                except Exception:
                    pass
            shutil.copy2(src, dst)
            print(f"📤 Copied → {dst}")
            return True
        except PermissionError as e:
            print(f"Attempt {i}/{retries}: PermissionError copying to {dst}: {e}")
        except Exception as e:
            print(f"Attempt {i}/{retries}: Error copying to {dst}: {e}")
        time.sleep(delay)
    return False

ok_pq = copy_with_retries(LOCAL_PARQUET, OUT_PARQUET)
ok_csv = copy_with_retries(LOCAL_SAMPLE, OUT_SAMPLE)
print(f"Copy parquet ok: {ok_pq} | Copy sample ok: {ok_csv}")
```

```python
# Quick validity checks for merged dataset (schema + metadata + spot checks)
import os
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import pyarrow.dataset as ds
from IPython.display import display

X_PQ = r"X:\\data\\processed\\nyc_2024_trips_weather.parquet"
LOCAL_PQ = os.path.abspath(os.path.join("..","data","processed","nyc_2024_trips_weather.parquet"))

def parquet_meta(path):
    pf = pq.ParquetFile(path)
    names = pf.schema_arrow.names
    meta = {
        "path": path,
        "exists": os.path.exists(path),
        "num_rows": pf.metadata.num_rows,
        "num_row_groups": pf.metadata.num_row_groups,
        "num_columns": pf.metadata.num_columns,
        "columns": names,
        "created_by": pf.metadata.created_by,
    }
    return meta

def print_meta(label, meta):
    print(f"[{label}] -> {meta['path']}")
    print(f"  exists={meta['exists']} | rows={meta['num_rows']:,} | row_groups={meta['num_row_groups']} | columns={meta['num_columns']}")
    print(f"  first 10 cols: {meta['columns'][:10]}")
    if meta.get("created_by"):
        print(f"  created_by: {meta['created_by']}")
    print()

# 1) Parquet metadata check (no full data load)
meta_x = parquet_meta(X_PQ)
print_meta("X drive", meta_x)
if os.path.exists(LOCAL_PQ):
    meta_local = parquet_meta(LOCAL_PQ)
    print_meta("Local", meta_local)
    print("Row count equal (X vs Local):", meta_x["num_rows"] == meta_local["num_rows"])
    print("Column names equal (X vs Local):", set(meta_x["columns"]) == set(meta_local["columns"]))
else:
    print("Local parquet not found (skipping local comparison)")

# 2) Required columns presence
required_cols = {
    "pickup_hour_local","PULocationID","trip_distance","fare_amount","tip_amount","total_amount","payment_type",
    "temp_c","rh_percent","wind_ms","precip_mm","visibility_km","weather_code","merge_status",
}
missing = sorted(list(required_cols - set(meta_x["columns"])))
print("Missing required columns:", missing)

# 3) Spot-check first row group only (lightweight), consistency of merge_status vs weather nullness
pf = pq.ParquetFile(X_PQ)
cols_probe = [
    "pickup_hour_local","merge_status","temp_c","rh_percent","wind_ms","precip_mm","visibility_km","weather_code",
    "PULocationID","trip_distance","fare_amount","tip_amount","total_amount","payment_type",
 ]
rg0 = pf.read_row_group(0, columns=[c for c in cols_probe if c in meta_x["columns"]])
df_rg0 = rg0.to_pandas()
print(f"Row-group 0 sample -> rows={len(df_rg0):,}")
if "merge_status" in df_rg0.columns:
    vc = df_rg0["merge_status"].value_counts(dropna=False)
    print("merge_status (row-group 0):\n", vc)
    weather_cols = [c for c in ["temp_c","rh_percent","wind_ms","precip_mm","visibility_km","weather_code"] if c in df_rg0]
    if len(weather_cols):
        matched = df_rg0["merge_status"] == "matched"
        no_w = df_rg0["merge_status"] == "no_weather_for_hour"
        matched_all_nonnull = df_rg0.loc[matched, weather_cols].notna().all(axis=1).mean() if matched.any() else float("nan")
        now_all_null = df_rg0.loc[no_w, weather_cols].isna().all(axis=1).mean() if no_w.any() else float("nan")
        print(f"matched rows: {matched.sum():,} | share all weather non-null: {matched_all_nonnull:.3f}")
        print(f"no_weather_for_hour rows: {no_w.sum():,} | share all weather null: {now_all_null:.3f}")
else:
    print("merge_status column not found in row-group 0 sample")

# 4) Dtype sanity: pickup_hour_local should be datetime64[ns] in the sample
if "pickup_hour_local" in df_rg0.columns:
    print("pickup_hour_local dtype:", df_rg0["pickup_hour_local"].dtype)
    if str(df_rg0["pickup_hour_local"].dtype).startswith("datetime64"):
        print("✅ pickup_hour_local is datetime64[ns] (sample)")
    else:
        print("⚠ pickup_hour_local is not datetime64 (sample)")
```

This step samples several hours across the year and verifies numeric weather equality (processed vs NOAA).

```python
# Verify weather alignment for a few sampled hours (early/mid/late)
import numpy as np, pandas as pd, pyarrow.parquet as pq
from datetime import timedelta
PROC_PQ = r"X:\\data\\processed\\nyc_2024_trips_weather.parquet"
NOAA_PQ = r"X:\\data\\interim\\noaa_hourly_citywide_2024.parquet"
pf = pq.ParquetFile(PROC_PQ)
# sample: pick one hour from RG0 and one from the last RG; also try mid RG if available
samples = []
def pick_one_hour_from_rg(pf, rg_index):
    tbl = pf.read_row_group(rg_index, columns=["pickup_hour_local"])
    s = pd.to_datetime(tbl.to_pandas()["pickup_hour_local"])
    if len(s)==0:
        return None
    # choose median row to avoid extreme skew
    return s.iloc[len(s)//2]
first_hour = pick_one_hour_from_rg(pf, 0)
last_hour = pick_one_hour_from_rg(pf, max(0, pf.metadata.num_row_groups-1))
mid_rg = pf.metadata.num_row_groups//2
mid_hour = pick_one_hour_from_rg(pf, mid_rg) if pf.metadata.num_row_groups>2 else None
for h in [first_hour, mid_hour, last_hour]:
    if h is not None:
        samples.append(h)
print("Sampled hours:", [str(h) for h in samples])
# build small frames around these hours from processed file
def load_processed_for_hours(hours):
    out=[]
    for h in hours:
        # read row groups and filter in pandas (small number of hours; quick)
        for i in range(pf.metadata.num_row_groups):
            tbl = pf.read_row_group(i, columns=["pickup_hour_local","temp_c","rh_percent","wind_ms","precip_mm","visibility_km","weather_code","merge_status"])
            df = tbl.to_pandas()
            mask = pd.to_datetime(df["pickup_hour_local"]).isin([h])
            if mask.any():
                out.append(df.loc[mask])
    return pd.concat(out, ignore_index=True) if out else pd.DataFrame()
proc_sample = load_processed_for_hours(samples)
print(f"Processed rows for sampled hours: {len(proc_sample):,}")
# load NOAA hourly and compare values for those hours
noaa = pd.read_parquet(NOAA_PQ)
noaa.rename(columns={"observation_hour_local":"pickup_hour_local"}, inplace=True)
noaa_sub = noaa[noaa["pickup_hour_local"].isin(samples)].copy()
# aggregate processed by hour (weather columns should be constant per hour)
weather_cols=["temp_c","rh_percent","wind_ms","precip_mm","visibility_km","weather_code"]
agg_proc = (proc_sample.groupby("pickup_hour_local")[weather_cols].agg({c:"nunique" for c in weather_cols}).rename(columns=lambda c:f"{c}_nuniq").reset_index())
print("Distinct weather values per hour in processed sample:\n", agg_proc)
# join and check equality
merged_chk = proc_sample.merge(noaa_sub, on="pickup_hour_local", suffixes=("_proc","_noaa"), how="left")
eq = {}
for c in weather_cols:
    a = merged_chk[f"{c}_proc"]
    b = merged_chk[f"{c}_noaa"]
    eq[c] = float((a.fillna(np.nan)==b.fillna(np.nan)).mean())
print("Share of rows equal for each weather column (proc vs NOAA):", eq)
```

This step scans all row groups to confirm that rows labeled 'no_weather_for_hour' have weather columns null, and 'matched' rows have weather populated where available.

```python
# Global consistency: scan row groups for weather-null logic by merge_status
import numpy as np, pyarrow.parquet as pq
PROC_PQ = r"X:\\data\\processed\\nyc_2024_trips_weather.parquet"
pf = pq.ParquetFile(PROC_PQ)
weather_cols=["temp_c","rh_percent","wind_ms","precip_mm","visibility_km","weather_code"]
tot=0; no_w_tot=0; no_w_allnull=0; matched_tot=0; matched_allnonnull=0
for i in range(pf.metadata.num_row_groups):
    tbl = pf.read_row_group(i, columns=["merge_status"]+weather_cols)
    df = tbl.to_pandas()
    tot += len(df)
    m_no = df["merge_status"]=="no_weather_for_hour"
    m_ok = df["merge_status"]=="matched"
    no_w_tot += m_no.sum()
    matched_tot += m_ok.sum()
    no_w_allnull += df.loc[m_no, weather_cols].isna().all(axis=1).sum()
    matched_allnonnull += df.loc[m_ok, weather_cols].notna().all(axis=1).sum()
print(f"Rows total: {tot:,}")
print(f"no_weather_for_hour rows: {no_w_tot:,} | share all weather null: {no_w_allnull/max(1,no_w_tot):.3f}")
print(f"matched rows: {matched_tot:,} | share all weather non-null: {matched_allnonnull/max(1,matched_tot):.3f}")
```

#### Processed vs NOAA by hour — compact report
This step aggregates processed trips per hour and compares numeric weather columns to NOAA across 2024.

```python
# Build hour-level report: processed (X: processed parquet) vs NOAA (X: interim parquet)
import os, numpy as np, pandas as pd, pyarrow.parquet as pq
from math import isfinite
PROC_PQ = r"X:\\data\\processed\\nyc_2024_trips_weather.parquet"
NOAA_PQ = r"X:\\data\\interim\\noaa_hourly_citywide_2024.parquet"
numeric_cols = ["temp_c","rh_percent","wind_ms","precip_mm","visibility_km"]
# --- Load NOAA (small) ---
noaa = pd.read_parquet(NOAA_PQ)
noaa = noaa.rename(columns={"observation_hour_local":"hour"})[["hour"]+numeric_cols].copy()
noaa_hours = set(noaa["hour"].astype("datetime64[ns]").tolist())
# --- Scan processed parquet by row group; aggregate per hour ---
pf = pq.ParquetFile(PROC_PQ)
parts = []
for i in range(pf.metadata.num_row_groups):
    tbl = pf.read_row_group(i, columns=["pickup_hour_local"]+numeric_cols)
    df = tbl.to_pandas()
    df = df.rename(columns={"pickup_hour_local":"hour"})
    # Mean is safe because hour-level weather is constant for a given hour; NaNs are ignored
    part = df.groupby("hour", as_index=False)[numeric_cols].mean()
    parts.append(part)
proc = pd.concat(parts, ignore_index=True)
proc = proc.groupby("hour", as_index=False)[numeric_cols].mean()  # reconcile across row groups
proc_hours = set(proc["hour"].astype("datetime64[ns]").tolist())
# --- Coverage ---
missing_in_proc = sorted(list(noaa_hours - proc_hours))
missing_in_noaa = sorted(list(proc_hours - noaa_hours))
print("Coverage:")
print(f"  NOAA hours:       {len(noaa_hours):,}")
print(f"  Processed hours:  {len(proc_hours):,}")
print(f"  Intersection:     {len(noaa_hours & proc_hours):,}")
print(f"  Missing in proc:  {len(missing_in_proc)}", ("→ e.g.," , missing_in_proc[:3]) if missing_in_proc[:3] else "")
print(f"  Missing in NOAA:  {len(missing_in_noaa)}", ("→ e.g.," , missing_in_noaa[:3]) if missing_in_noaa[:3] else "")
# --- Join and compute equality metrics ---
merged = noaa.merge(proc, on="hour", how="inner", suffixes=("_noaa","_proc"))
report = {}
for c in numeric_cols:
    a = merged[f"{c}_noaa"].to_numpy()
    b = merged[f"{c}_proc"].to_numpy()
    eq = np.isclose(a, b, equal_nan=True, atol=1e-9)
    share = float(np.mean(eq)) if len(eq) else float("nan")
    # max absolute diff over valid comparisons
    diffs = np.abs(a - b)
    diffs = diffs[np.isfinite(diffs)] if len(diffs) else diffs
    maxdiff = float(diffs.max()) if len(diffs) else float("nan")
    report[c] = {"share_equal": share, "max_abs_diff": maxdiff}
print("\nEquality by numeric column (processed vs NOAA over intersection):")
for c, stats in report.items():
    print(f"  {c:14} share_equal={stats['share_equal']:.6f}  max_abs_diff={stats['max_abs_diff']:.3g}")
# --- Verdict ---
all_equal = all((np.isnan(v["share_equal"]) or v["share_equal"]==1.0) for v in report.values())
print("\nVerdict:")
print("  Numeric weather columns match NOAA per hour:" , all_equal)
if not all_equal:
    print("  Note: small float differences or NaNs in NOAA can reduce equality; see max_abs_diff above.")
```
